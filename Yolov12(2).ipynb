```python
from ultralytics import YOLO
model = YOLO('yolov12n.pt')

# Perform object detection on an image
results = model("/content/drive/MyDrive/Cotton-Disease-16/test/images/102_jpg.rf.3481412c6aa24ac3bf269421e48c1b0b.jpg")

results[0].show()
```

```python
from ultralytics import YOLO

model = YOLO('yolov12n.yaml')

# Train the model
results = model.train(
  data='/content/drive/MyDrive/Cotton-Disease-16/data.yaml',
  epochs=15,
  batch=32,
  imgsz=640,
  scale=0.4,  # S:0.9; M:0.9; L:0.9; X:0.9
  mosaic=1.0,
  mixup=0.0,  # S:0.05; M:0.15; L:0.15; X:0.2
  copy_paste=0.1,  # S:0.15; M:0.4; L:0.5; X:0.6
  device="0",
)
```
